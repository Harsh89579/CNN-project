```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv("C:\\Users\\harsh\\OneDrive\\Documents\\churn dataset1.csv")
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
input_data = dataset.iloc[:,:-1]
output_data = dataset.iloc[:,-1]
```

```python
le = LabelEncoder()
output_data = le.fit_transform(output_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
```

```python
input_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, y_train.shape)
```

```python
import tensorflow
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6,input_dim= 8,activation="relu"))
ann.add(Dense(4,activation="relu"))
ann.add(Dense(2,activation="relu"))
ann.add(Dense(1,activation="sigmoid"))
```

```python
ann.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train, batch_size=1,epochs= 20,callbacks=EarlyStopping())
```

```python
prd1= ann.predict(x_train)
prd_data1=[]
for i in prd1:
    if i[0]>0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```

```python
prd = ann.predict(x_test)
```

```python
prd_data=[]
for i in prd:
    if i[0]>0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
    
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data)*100
```

```python
accuracy_score(y_train,prd_data1)*100
```
